# Plotting Numpy Arrays 

##### Data visualization is the representation of data in a graphical form , which makes it easier to understand and analyze large and complex datasets.
##### Matplotlib and Seaborn are two popular libraries for data visualization in Python , and they Provide a comprehensive set of tools for creating and exploring visualization of various types. 

**1. Matplotlib** it is a widely-used library for data visualization in python . It provides a comprehensive set of tools for creating visualizations  of various types , and it integrates well with Numpy , making it easy to work with arrays of numerical data.

**2 Seaborn :** It provides an easy-to-use interfave for creating visually appealing and informative plots , and it includes advanced visualization techniques such as heat maps , violin plots and pair plots . Seacborn also provides built in themes and color palettes that make it easier to produce aesthetically pleasing plots with a minimum amount of efforts

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sb
```

```python
import numpy as np
```


# Matplotlib

```python
#line plot 
x= np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('line-plot')
plt.show()
```

```python
#scatter plot - dispay the relationship b/w 2 variable as individual data points which are represented by dots

x = np.random.rand(100)
#Genrates the random array with 100 elements
y = np.random.rand(100)

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
```

```python
#histogram - bar plot shows the distribution of continuous data
data = np.random.normal(100,20,1000)
#height is basiclly the frequency of the bar
#bins is the no of bars in your graph  
plt.hist(data , bins=20,color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

```python
#barplot

x = ['A','B','C','D']
y = [ 1,5,2,8]
plt.bar(x,y)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
labels= ['A','B','C','D']
values = [ 10,20,30,40]

plt.pie(values , labels = labels , startangle=90 , autopct='%1.1f%%')
#autopct is basically used to show the values in percentage
plt.axis('equal')
#plt.axis('equal') means x and y axis are scaled equal
plt.title('Bar Plot')
plt.show()
```

```python
#box plot = this plot display the summary for the datavalues like maximum , minimum , the first quantile ,median , third quantile 

data = [np.random.normal(50,10,100) , np.random.normal(80,20,100) , np.random.normal(40,15,100)]

plt.boxplot(data)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

# Seaborn

```python
#line plot

x=np.linspace(0,10,100)
y=np.sin(x)

# Plot the data 
sb.lineplot(y)

# show the plot
plt.show()
```

```python
 #pair plot 
import pandas as pd 
#Generating Random data
data = np.random.randn(100,5)
df = pd.DataFrame(data , columns = ['A','B','C','D','E'])

#plotting a pair plot 
sb.pairplot(df)
#Adding title to the plot 

plt.suptitle("Pair Plot of the Data")

#showing the plot

plt.show()
```

```python
#heat map 
data = np.random.random((10,10))

#plot the data
sb.heatmap(data)

#show the plot
plt.show()
```

```python
#violin plot = combination of box plot and corlonel density plot
data= np.random.normal(0,1,100)
#plot the data 
sb.violinplot(data)

#show the plot 
plt.show()
```
